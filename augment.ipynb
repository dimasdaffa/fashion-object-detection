```python
import cv2
import os
import numpy as np

# Path ke folder sumber dan output
input_folder = r"D:\fashion-object-detection\fashion_mine\watches"
output_folder = r"D:\fashion-object-detection\fashion_mine\watches_blurred"

# Membuat folder output jika belum ada
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop untuk semua file dalam folder
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):  # Filter hanya file gambar
        img_path = os.path.join(input_folder, filename)
        img = cv2.imread(img_path)

        if img is not None:
            # Pilih sigma kecil secara random antara 0.01 dan 0.03
            sigma = np.random.uniform(0.01, 0.03)
            kernel_size = (5, 5)  # Kernel kecil tetap dipakai untuk sigma kecil

            # Terapkan Gaussian blur
            blurred_img = cv2.GaussianBlur(img, kernel_size, sigma)

            # Simpan hasil ke folder output
            output_path = os.path.join(output_folder, filename)
            cv2.imwrite(output_path, blurred_img)
            print(f"Processed: {filename}")

print("Augmentasi selesai.")
```

```python
import cv2
import os
import numpy as np

# Path ke folder sumber dan output
input_folder = r"D:\fashion-object-detection\fashion_mine\shoes"
output_folder = r"D:\fashion-object-detection\fashion_mine\shoes_zoom_out"

# Membuat folder output jika belum ada
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop untuk semua file dalam folder
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):  # Filter hanya file gambar
        img_path = os.path.join(input_folder, filename)
        img = cv2.imread(img_path)

        if img is not None:
            h, w, _ = img.shape
            zoom_out_ratio = 1.5  # 150% zoom out

            # Hitung ukuran frame baru dengan zoom out
            new_h = int(h * (1 + zoom_out_ratio))
            new_w = int(w * (1 + zoom_out_ratio))

            # Buat frame putih dengan ukuran baru
            frame = np.ones((new_h, new_w, 3), dtype=np.uint8) * 255

            # Tempelkan gambar asli di tengah frame
            y_offset = (new_h - h) // 2
            x_offset = (new_w - w) // 2
            frame[y_offset:y_offset + h, x_offset:x_offset + w] = img

            # Simpan hasil ke folder output
            output_path = os.path.join(output_folder, filename)
            cv2.imwrite(output_path, frame)
            print(f"Processed: {filename}")

print("Augmentasi zoom out selesai.")
```

```python
import cv2
import os
import numpy as np

# Path ke folder sumber dan output
input_folder = r"D:\fashion-object-detection\fashion_mine\shirts"
output_folder = r"D:\fashion-object-detection\fashion_mine\shirts_brightness_down"

# Membuat folder output jika belum ada
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop untuk semua file dalam folder
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):  # Filter hanya file gambar
        img_path = os.path.join(input_folder, filename)
        img = cv2.imread(img_path)

        if img is not None:
            # Pilih faktor brightness secara random antara 0.5 hingga 0.8 (penurunan brightness)
            brightness_factor = np.random.uniform(0.5, 0.8)

            # Terapkan penurunan brightness
            img_bright = np.clip(img * brightness_factor, 0, 255).astype(np.uint8)

            # Simpan hasil ke folder output
            output_path = os.path.join(output_folder, filename)
            cv2.imwrite(output_path, img_bright)
            print(f"Processed: {filename}")

print("Penurunan brightness selesai.")
```

```python
import os
import shutil

# Path folder sumber dan folder tujuan
source_folder = r"D:\fashion-object-detection\fashion_mine"
destination_folder = r"D:\fashion-object-detection\fashion_mine\all_fashion"

# Membuat folder tujuan jika belum ada
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# Loop melalui semua subfolder dan file dalam folder sumber
for root, _, files in os.walk(source_folder):
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):  # Filter hanya file gambar
            source_path = os.path.join(root, file)
            destination_path = os.path.join(destination_folder, file)

            # Pastikan tidak ada duplikasi nama file di folder tujuan
            counter = 1
            while os.path.exists(destination_path):
                name, ext = os.path.splitext(file)
                destination_path = os.path.join(destination_folder, f"{name}_{counter}{ext}")
                counter += 1

            # Pindahkan file ke folder tujuan
            shutil.move(source_path, destination_path)
            print(f"Moved: {source_path} -> {destination_path}")

print("Semua gambar telah dipindahkan ke folder all_fashion.")
```
